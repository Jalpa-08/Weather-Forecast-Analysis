```python
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
```

```python
plt.figure(figsize=(20, 10))
plt.bar(data["Date/Time"], data["Temp_C"], color='blue')
plt.xlabel('Date/Time', fontsize=15, color='r')
plt.ylabel('Temp_C', fontsize=15, color='r')
plt.title('temperatures vs date/time', fontsize=20, color='k')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(data.index, data['Press_kPa'],  color='g', label= 'pressure kpa')
plt.plot(data.index, data['Rel Hum_%'], color='c',label= 'relative humadity')
plt.plot(data.index, data['Wind Speed_km/h'],  color='k', label= 'speed of wind')
plt.legend()
plt.show()
```
